```python
import pandas as pd
import glob
import os
import re
```

# Find and list all files we want to compile

```python
input_dir = "/projects/ps-yeolab4/NCRCRG/home/bay001/results_dir/"
all_metrics_summaries = glob.glob(os.path.join(input_dir, "*/*/*/outs/metrics_summary.csv"))
all_metrics_summaries
```

# Use read pd.concat() to merge each dataframe into one

```python
regex = "^\/projects\/ps-yeolab4\/NCRCRG\/home\/bay001\/results_dir\/([\w\d\s]+)\/results\/[\w\d\/.]+$"
regex = re.compile(regex)

merged = pd.DataFrame()
for summary in all_metrics_summaries:
    df = pd.read_csv(summary)
    df.index = [re.match(regex, summary).group(1)]
    merged = pd.concat([merged, df])

merged
```

# Calculate some secondary statistics

```python
def str2num(string):
    """ converts a comma-formatted string into an integer type. """
    return int(string.replace(',',''))

def str2perc(string):
    """ converts a percentage string into an float type. """
    return float(string.replace('%',''))*.01

merged['For 25k reads/cell'] = merged.apply(lambda x: str2num(x['Estimated Number of Cells'])*25000., axis=1)
merged['For 50k reads/cell'] = merged.apply(lambda x: str2num(x['Estimated Number of Cells'])*50000., axis=1)
merged['Effective reads/cell'] = merged.apply(lambda x: str2num(x['Mean Reads per Cell'])*str2perc(x['Fraction Reads in Cells']), axis=1)
```
